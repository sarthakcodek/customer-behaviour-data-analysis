```python
d['Review Rating'] = d.groupby('Category')['Review Rating'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
d.columns = d.columns.str.lower()
d.columns = d.columns.str.replace(" ", '_')
```

```python
d=d.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
labels = ['Young-adult', 'Adult', 'Mid-aged', 'Senior']
d['age_group'] = pd.qcut(d['age'], q=4, labels=labels)
```

```python
freq_map = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
```

```python
d['purchase_frequency_days'] = d['frequency_of_purchases'].map(freq_map)
```
